```python
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Path to your local HTML file
html_path = r"/content/Full text of _ಅರೆಭಾಷೆ ಮುಡ್ಪು (Paragalda Tinga)_.html"
```

```python
# Read and parse the HTML
with open(html_path, "r", encoding="utf-8") as file:
    html_content = file.read()
```

```python
soup = BeautifulSoup(html_content, "html.parser")
text = soup.get_text(separator="\n", strip=True)
```

```python
sentences = [line.strip() for line in re.split(r'[.!?…\n]', text) if line.strip()]
label = "Arre bhashe"
labeled_data = [{"text": sent, "label": label} for sent in sentences]
```

```python
df = pd.DataFrame(labeled_data)
df.to_csv("dialect_labeled_text.csv", index=False, encoding='utf-8-sig')

print(f"✅ Saved {len(df)} labeled sentences.")
```
