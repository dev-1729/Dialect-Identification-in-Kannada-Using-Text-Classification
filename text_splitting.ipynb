```python
import pandas as pd

# Load your CSV
df = pd.read_csv('kannada_subtitles.csv')

# Parameters
max_words = 10

# Function to split text into word-limited chunks
def split_text_by_max_words_basic(text, max_words):
    if not isinstance(text, str):
        return []
    words = text.split()
    return [' '.join(words[i:i+max_words]) for i in range(0, len(words), max_words)]

# Apply the function
df['Chunks'] = df['text'].apply(lambda x: split_text_by_max_words_basic(x, max_words))

# Explode and keep only required columns
df_exploded = df.explode('Chunks').reset_index(drop=True)
df_final = df_exploded[['Chunks', 'label']].rename(columns={'Chunks': 'text', 'label': 'dialect'})

# Save to CSV
df_final.to_csv('processed_kannada_data.csv', index=False, encoding='utf-8-sig')
```
